```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('commodity_index.csv')
```

```python
df.head()
```

```python
df.commodity.describe()
```

```python
CI = df.commodity.to_numpy()
```

```python
np.std(CI)
```

```python
df.plot(x='date', y='commodity')
```

```python
from pandas.plotting import autocorrelation_plot
```

```python
autocorrelation_plot(df.commodity)
```

```python
r = np.log(CI[1:]) - np.log(CI[:-1])
```

```python
print(np.mean(r), np.std(r))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.displot(r)
```

```python
sns.boxplot(r)
```

```python
import statsmodels.api as sm
```

```python
sm.qqplot(r, scale=0.1, fit=True, line='45')
```

```python
import scipy.stats as stats
```

```python
sm.qqplot(r, stats.t, distargs=(10,), fit=True, line='45')
```
